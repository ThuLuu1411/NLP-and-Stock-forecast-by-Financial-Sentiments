```python
from sklearn.feature_extraction.text import CountVectorizer
X = df['title_new'].astype('str')
vectorizer = CountVectorizer()
vectorizer.fit(X)
X_vec = vectorizer.transform(X)
tfidf_feature_names = vectorizer.get_feature_names_out()
sentiments = pd.DataFrame(data=X_vec.toarray(), columns = tfidf_feature_names)
dummies = pd.get_dummies(df['ma_ck'])
sentiments[dummies.columns] = dummies
sentiments
```

```python
sentiments_columns = pd.read_csv('Preprocessing/20230205/sentiments_vector.csv').columns.tolist()[:337]
columns = [i for i in sentiments.columns if i in sentiments_columns]
sentiments = sentiments[columns]
```

```python
import pickle
model = pickle.load(open('data/finalized_model.sav', 'rb'))
```

```python
df['sentiment_negative'] = model.predict(sentiments)
df['sentiment_negative_prob'] = model.predict_proba(sentiments.values)[:,1]
df
```

```python
df[['date', 'ma_ck', 'title', 'sentiment_negative', 'sentiment_negative_prob']].to_csv('data/sentiment_score.csv', index=False)
```
