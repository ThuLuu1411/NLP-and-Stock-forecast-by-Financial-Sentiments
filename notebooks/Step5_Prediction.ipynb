```python
# Nạp các gói thư viện NumPy và Pandas
import pandas as pd
import numpy as np
import regex
import string
import pickle
```

```python
#Tạo đường link
import sys
import os

current = os.path.abspath('')
parent = os.path.dirname(current)
grandparent = os.path.dirname(parent)
```

```python
#File path

#Model
model_file = parent+"/data/processed/regression_model.sav"

#Media data
input_file = parent+"/data/processed/input_table.csv"

## File stock data
price_file = parent+'/data/processed/price07_01_2023.xlsx'
```

# Dự báo return

```python
model = pickle.load(open(model_file, 'rb'))
```

```python
model.feature_names_in_
```

```python
data = pd.read_csv(input_file)
data
```

```python
data['Date'] = pd.DatetimeIndex(data['new_date'])
data = data.reset_index(drop=True)
data
```

```python
dummies = pd.get_dummies(data['ma_ck'])
dummies
from sklearn.feature_extraction.text import CountVectorizer
X = data['title_new']
vectorizer = CountVectorizer()
vectorizer.fit(X)
X_vec = vectorizer.transform(X)
tfidf_feature_names = vectorizer.get_feature_names_out()
sentiments = pd.DataFrame(data=X_vec.toarray(), columns = tfidf_feature_names)
sentiments[dummies.columns] = dummies
sentiments
```

```python
sentiments_df = pd.DataFrame(index=sentiments.index)
for i in model.feature_names_in_:
  if i in sentiments.columns:
    sentiments_df[i]=sentiments[i]
  else:
    sentiments_df[i]=0
sentiments_df
```

```python
y_pred = model.predict(sentiments_df)
```

```python
y_pred
```

```python
price_output = data[['ma_ck', 'Date', 'session']]
price_output['pred_return'] = y_pred
price_output = price_output[price_output['session']=='return_in_session']
price_output
```

# Du bao gia

```python
price_root = pd.read_excel(price_file, usecols = ['Date', 'maCK', 'Open', 'Close'])
price = price_root.copy()
price['Date'] = pd.DatetimeIndex(price['Date'], dayfirst=True)
price = price.rename(columns={'Date':'date',
                              'maCK':'ma_ck',
                              'Open':'open_price',
                              'Close':'close_price'})
price = price.sort_values(['ma_ck', 'date'], ascending=True)
price['open_price'] = price['open_price'].astype('str')
price['open_price'] = price['open_price'].str.replace('00.0','000')
price['open_price'] = price['open_price'].str.replace(',', '.')
# price = price[price['open_price'].isnull()==False]
price = price.rename(columns={'date': 'Date'})
price['open_price'] = price['open_price'].astype('float', errors='raise')
price = price.drop_duplicates(ignore_index=True)
price
```

```python
price_df = price_output.merge(price, on=['Date', 'ma_ck'], how='left')
price_df['pred_price'] = np.power(10, price_df['pred_return'])
price_df['pred_price'] = price_df['pred_price']*price_df['open_price']
price_df
```

# AUTOREGRESSIVE

```python
price
```

```python
from statsmodels.tsa.ar_model import AutoReg
def ar(ma_ck):
  df = price[price['ma_ck']==ma_ck].sort_values('Date', ignore_index=True)
  pred_df = df[['Date', 'ma_ck']]
  reg = AutoReg(df['close_price'], 1)
  pred_df['pred_ar'] = reg.fit().predict()
  return pred_df
```

```python
ar_df = ar('ABC')
```

```python
for i in price_df['ma_ck'].unique():
  ar_df = ar_df.append(ar(i))
```

```python
ar_df
```

```python
price_df = price_df.merge(ar_df, on=['Date', 'ma_ck'], how='left')
price_df
```

# SO SANH AUTOREGRESSIVE VA NLP

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
def accuracy(ma_ck):
  df = price_df[price_df['ma_ck']==ma_ck]
  r2_nlp = r2_score(df['close_price'], df['pred_price'])
  r2_ar = r2_score(df['close_price'], df['pred_ar'])
  mse_nlp = mean_squared_error(df['close_price'], df['pred_price'])
  mse_ar = mean_squared_error(df['close_price'], df['pred_ar'])
  row = {'ma_ck':ma_ck, 'r2_nlp':r2_nlp, 'r2_ar':r2_ar, 'mse_nlp':mse_nlp, 'mse_ar':mse_ar}
  return row
```

```python
mean_squared_error(price_df['close_price'], price_df['pred_price'])
```

```python
r2_score(price_df['close_price'], price_df['pred_ar'])
```

```python
compare_table = pd.DataFrame(columns=['ma_ck', 'r2_nlp', 'r2_ar', 'mse_nlp', 'mse_ar'])
```

```python
for i in price_df['ma_ck'].unique():
  compare_table = compare_table.append(accuracy(i), ignore_index=True)
compare_table
```
