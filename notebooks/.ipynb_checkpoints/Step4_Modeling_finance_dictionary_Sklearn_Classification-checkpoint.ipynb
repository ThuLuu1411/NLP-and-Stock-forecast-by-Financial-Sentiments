```python
# Split train and test sample
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(sentiments_scaled, y, test_size=0.2, random_state=42)
X_train = X
X_test = X
y_train = y
y_test = y
```

```python
#FITTING THE CLASSIFICATION MODEL using Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, f1_score, accuracy_score, confusion_matrix,roc_curve,auc
model=RandomForestClassifier(random_state = 42)
model.fit(X_train, y_train.values)

#Predict y value for test dataset
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]


print(classification_report(y_test,y_pred))
print('Confusion Matrix:',confusion_matrix(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
print('AUC:', roc_auc)
```

```python
data = pd.read_csv('Preprocessing/20230305/input_table.csv')
data['Date'] = pd.DatetimeIndex(data['new_date'])
data = data[data['Date']<'2023-02-15'].reset_index(drop=True)
data = data[data['return']!=0]
data['return'] = data['return'].apply(get_binary)
dummies = pd.get_dummies(data['ma_ck'])
dummies
from sklearn.feature_extraction.text import CountVectorizer
X = data['title_new']
vectorizer = CountVectorizer()
vectorizer.fit(X)
X_vec = vectorizer.transform(X)
tfidf_feature_names = vectorizer.get_feature_names_out()
sentiments = pd.DataFrame(data=X_vec.toarray(), columns = tfidf_feature_names)
sentiments[dummies.columns] = dummies
sentiments_df = pd.DataFrame(index=sentiments.index)
for i in model.feature_names_in_:
  if i in sentiments.columns:
    sentiments_df[i]=sentiments[i]
  else:
    sentiments_df[i]=0
sentiments_df = sentiments_df.fillna(0)
```

```python
data['return'].value_counts()
```

```python
#Predict y value for test dataset
y_pred = model.predict(sentiments_df)
y_prob = model.predict_proba(sentiments_df)[:,1]


print(classification_report(data['return'],y_pred))
print('Confusion Matrix:',confusion_matrix(data['return'], y_pred))

fpr, tpr, thresholds = roc_curve(data['return'], y_prob)
roc_auc = auc(fpr, tpr)
print('AUC:', roc_auc)
```
